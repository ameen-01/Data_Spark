# Merging the data 

```python
df_merged=pd.merge(customers_df,sales_df,on='CustomerKey',how='inner')
```

```python
df_merged
```

```python
df_merged.isnull().sum()
```

```python
df_merged=pd.merge(df_merged,products_df,on='ProductKey',how='inner')
```

```python
df_merged=pd.merge(df_merged,stores_df,on='StoreKey',how='inner')
```

```python
df_merged=pd.merge(df_merged,exchange_df,left_on=['Order Date','Currency Code'],right_on=['Date','Currency'],how='inner')
```

```python
df_merged.info()
```

```python
df_merged.isna().sum()
```

```python
df_merged.columns
```

```python
df_merged.drop(columns=["State_y","Country_y"],inplace=True)
```

```python
df_merged.rename(columns={"State_x":"State","Country_x":"Country"},inplace=True)
```

```python
columnList=list(df_merged.columns)
```

```python
updated_columns=[]
```

```python
for i in columnList:
    updated_columns.append(i.replace(' ','_').lower())
```

```python
updated_columns
```

```python
df_merged.columns=updated_columns
```

```python
df_merged
```

```python
df_merged.isnull().sum()
```

```python
df_merged.drop_duplicates(inplace=True)
```

```python
op_path6="/Users/ameen/Documents/PROJECTS_NEW/DATA_SPARK/cleaned_data/merged_data.csv"
df_merged.to_csv(op_path6,index=False)
```

# Migrating To SQL

```python
import mysql.connector
```

```python
#establishing connection with sql server
from mysql import connector
connection = connector.connect(
    host='localhost',
    user='root',
    password=''
)

cursor=connection.cursor()
```

```python


#creating a dedicated DB for dataspark
cursor.execute('create database if not exists Data_Spark')

```

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#initiating engine
host='localhost'
user='root'
password=''
db='Data_Spark'
port='3306'
engine_q=f"mysql+mysqlconnector://{user}:{password}@{host}:{port}/{db}"
engine=create_engine(engine_q)
```

```python

df_merged.to_sql("merged_data",engine,if_exists='replace',index=False)

customers_df.to_sql("customers",engine,if_exists='replace',index=False)

sales_df.to_sql("sales",engine,if_exists='replace',index=False)

products_df.to_sql("products",engine,if_exists='replace',index=False)

stores_df.to_sql("stores",engine,if_exists='replace',index=False)

exchange_df.to_sql("exchange_rates",engine,if_exists='replace',index=False)

```
